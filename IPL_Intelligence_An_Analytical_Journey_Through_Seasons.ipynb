```python
deliveries['over'] = deliveries['ball'].astype(str).apply(lambda x: int(x.split('.')[0])) + 1

def determine_segment(over):
    if over <= 6:
        return 'Powerplay'
    elif 7 <= over <= 15:
        return 'Middle Overs'
    elif over >= 16:
        return 'Death Overs'
    return 'Unknown'

deliveries['segment'] = deliveries['over'].apply(determine_segment)

grouped = deliveries.groupby(['season', 'segment'])

total_runs = grouped['runs_off_bat'].sum() + grouped['extras'].sum()
total_overs = grouped['ball'].count() / 6
run_rate_per_segment = total_runs / total_overs

print("Run Rate Per Over for Each Season and Segment:")
print(run_rate_per_segment)



```

```python
import pandas as pd
import numpy as np

data = pd.read_csv('/content/drive/MyDrive/All fruits/match_data.csv', low_memory=False)

data['start_date'] = pd.to_datetime(data['start_date'])

data_2023 = data[data['start_date'].dt.year == 2023].copy()

data_2023['ball'] = data_2023['ball'].astype(str)
data_2023['over'] = data_2023['ball'].apply(lambda x: int(x.split('.')[0]))

powerplay = data_2023[data_2023['over'].between(1, 6)].copy()
middle_overs = data_2023[data_2023['over'].between(7, 15)].copy()
death_overs = data_2023[data_2023['over'].between(16, 20)].copy()

def calculate_economy(grouped_data):
    stats = grouped_data.agg({
        'runs_off_bat': 'sum',
        'extras': 'sum',
        'ball': 'count'
    })
    stats['overs'] = stats['ball'] / 6
    stats['total_runs_conceded'] = stats['runs_off_bat'] + stats['extras']
    stats['economy_rate'] = stats['total_runs_conceded'] / stats['overs']
    return stats

powerplay_economy = calculate_economy(powerplay.groupby('bowling_team'))
middle_overs_economy = calculate_economy(middle_overs.groupby('bowling_team'))
death_overs_economy = calculate_economy(death_overs.groupby('bowling_team'))

top_two_powerplay_economy_teams = powerplay_economy.sort_values(by='economy_rate', ascending=False).head(2).index.tolist()
top_two_middle_overs_economy_teams = middle_overs_economy.sort_values(by='economy_rate', ascending=False).head(2).index.tolist()
top_two_death_overs_economy_teams = death_overs_economy.sort_values(by='economy_rate', ascending=False).head(2).index.tolist()

print("Top two teams with the highest bowling economy in Powerplay:")
for i in top_two_powerplay_economy_teams:
  print(i)
print('\n')

print("Top two teams with the highest bowling economy in Middle Overs:")
for i in top_two_middle_overs_economy_teams:
  print(i)
print('\n')

print("Top two teams with the highest bowling economy in Death Overs:")
for i in top_two_death_overs_economy_teams:
  print(i)

```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = pd.read_csv('/content/drive/MyDrive/All fruits/match_data.csv', low_memory=False)
data['start_date'] = pd.to_datetime(data['start_date'])

data_2023 = data[data['start_date'].dt.year == 2023].copy()

data_2023['ball'] = data_2023['ball'].astype(str)
data_2023['over'] = data_2023['ball'].apply(lambda x: int(x.split('.')[0]))

team_name_mapping = {
    'Mumbai Indians': 'MI',
    'Chennai Super Kings': 'CSK',
    'Kolkata Knight Riders': 'KKR',
    'Royal Challengers Bangalore': 'RCB',
    'Sunrisers Hyderabad': 'SRH',
    'Delhi Capitals': 'DC',
    'Rajasthan Royals': 'RR',
    'Punjab Kings': 'PBKS',
    'Gujarat Titans': 'GT',
    'Lucknow Super Giants': 'LSG'
}

powerplay = data_2023[data_2023['over'].between(1, 6)].copy()
middle_overs = data_2023[data_2023['over'].between(7, 15)].copy()
death_overs = data_2023[data_2023['over'].between(16, 20)].copy()

powerplay.loc[:, 'bowling_team'] = powerplay['bowling_team'].map(team_name_mapping)
middle_overs.loc[:, 'bowling_team'] = middle_overs['bowling_team'].map(team_name_mapping)
death_overs.loc[:, 'bowling_team'] = death_overs['bowling_team'].map(team_name_mapping)

def calculate_metrics(phase_data):
    grouped = phase_data.groupby('bowling_team')

    total_runs = grouped['runs_off_bat'].sum()
    total_extras = grouped['extras'].sum()
    total_overs = grouped['ball'].count() / 6

    run_rate = total_runs / total_overs
    economy_rate = (total_runs + total_extras) / total_overs

    boundary_count = grouped['runs_off_bat'].apply(lambda x: (x == 4).sum() + (x == 6).sum())
    wickets = grouped['player_dismissed'].count()

    metrics = pd.DataFrame({
        'run_rate': run_rate,
        'boundary_count': boundary_count,
        'wickets': wickets,
        'economy_rate': economy_rate,
    })

    return metrics

powerplay_metrics = calculate_metrics(powerplay)
middle_overs_metrics = calculate_metrics(middle_overs)
death_overs_metrics = calculate_metrics(death_overs)

plt.figure(figsize=(10, 6))
sns.heatmap(data=powerplay_metrics[['economy_rate']], annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Economy Rate Heatmap in Powerplay (2023)")
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(data=middle_overs_metrics[['economy_rate']], annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Economy Rate Heatmap in Middle Overs (2023)")
plt.show()

plt.figure(figsize=(10, 6))
sns.heatmap(data=death_overs_metrics[['economy_rate']], annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Economy Rate Heatmap in Death Overs (2023)")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=powerplay_metrics, x='wickets', y='boundary_count', hue=powerplay_metrics.index)
plt.title("Boundary Count vs Wickets in Powerplay (2023)")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=middle_overs_metrics, x='wickets', y='boundary_count', hue=middle_overs_metrics.index)
plt.title("Boundary Count vs Wickets in Middle Overs (2023)")
plt.show()

plt.figure(figsize=(10, 6))
sns.scatterplot(data=death_overs_metrics, x='wickets', y='boundary_count', hue=death_overs_metrics.index)
plt.title("Boundary Count vs Wickets in Death Overs (2023)")
plt.show()

```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, export_text
from sklearn.metrics import accuracy_score, classification_report

data = pd.read_csv('/content/drive/MyDrive/All fruits/match_data.csv')

data_2023 = data.loc[data['season'] == 2023].copy()

data_2023.loc[:, 'over'] = data_2023['ball'].astype(str).apply(lambda x: int(x.split('.')[0]))

grouped_data = data_2023.groupby(['cricsheet_id', 'innings'], as_index=False).agg({
    'runs_off_bat': 'sum',
    'extras': 'sum',
    'season': 'first',
    'venue': 'first',
    'batting_team': 'first',
    'bowling_team': 'first'
})

first_innings_data = grouped_data.loc[grouped_data['innings'] == 1].copy()

first_innings_data.loc[:, 'total_score'] = first_innings_data['runs_off_bat'] + first_innings_data['extras']

score_bins = [70, 120, 200, float('inf')]
score_labels = ['Low', 'Medium', 'High']

first_innings_data.loc[:, 'score_category'] = pd.cut(
    first_innings_data['total_score'], bins=score_bins, labels=score_labels)

y = first_innings_data['score_category']

X = first_innings_data[['season', 'venue', 'batting_team', 'bowling_team']]

X_encoded = pd.get_dummies(X, drop_first=True)

X_train, X_test, y_train, y_test = train_test_split(X_encoded, y, test_size=0.2, random_state=42)

tree_classifier = DecisionTreeClassifier(max_depth=3)
tree_classifier.fit(X_train, y_train)

y_pred = tree_classifier.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
classification_rep = classification_report(y_test, y_pred)

print(f"Accuracy Score: {accuracy:.2%}")
print("Classification Report:")
print(classification_rep)

tree_structure = export_text(tree_classifier, feature_names=list(X_encoded.columns))
print("Decision Tree Structure:")
print(tree_structure)

```
